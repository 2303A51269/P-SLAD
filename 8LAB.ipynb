```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

study_hours = np.array([2, 4, 6, 8, 1, 5, 7, 3, 9, 10])
exam_scores = np.array([50, 60, 70, 85, 40, 65, 78, 55, 90, 95])
sleep_hours = np.array([8, 7, 6, 5, 9, 6.5, 5.5, 7.5, 4, 3])
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
sns.scatterplot(x=study_hours, y=exam_scores)
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.title("Study Hours vs Exam Scores")
plt.subplot(1, 3, 2)
sns.scatterplot(x=sleep_hours, y=exam_scores)
plt.xlabel("Sleep Hours")
plt.ylabel("Exam Scores")
plt.title("Sleep Hours vs Exam Scores")
plt.subplot(1, 3, 3)
sns.scatterplot(x=study_hours, y=sleep_hours)
plt.xlabel("Study Hours")
plt.ylabel("Sleep Hours")
plt.title("Study Hours vs Sleep Hours")
plt.tight_layout()
plt.show()
corr_study_exam, _ = pearsonr(study_hours, exam_scores)
corr_sleep_exam, _ = pearsonr(sleep_hours, exam_scores)
corr_study_sleep, _ = pearsonr(study_hours, sleep_hours)
print(f"Correlation between Study Hours and Exam Scores: {corr_study_exam:.2f}")
print(f"Correlation between Sleep Hours and Exam Scores: {corr_sleep_exam:.2f}")
print(f"Correlation between Study Hours and Sleep Hours: {corr_study_sleep:.2f}")
if corr_study_exam > 0.7:
    print("There is a strong positive correlation between study hours and exam scores.")
elif corr_study_exam > 0.3:
    print("There is a moderate positive correlation between study hours and exam scores.")
else:
    print("There is a weak or no correlation between study hours and exam scores.")

if corr_sleep_exam < -0.3:
    print("There is a negative correlation between sleep hours and exam scores, suggesting less sleep may be associated with higher scores.")

if corr_study_sleep < -0.3:
    print("There is a negative correlation between study hours and sleep hours, meaning more study hours may lead to less sleep.")
```
